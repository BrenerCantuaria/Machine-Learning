# Função de 2 imagens

```python
def plot_2img(img1,tit1,img2,tit2,tam):
    plt.figure(figsize = tam)
    
    plt.subplot(1,2,1)
    plt.imshow(img1)
    plt.title(tit1)
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(1,2,2)
    plt.imshow(img2)
    plt.title(tit2)
    plt.xticks([]), plt.yticks([])
    
    plt.show()
```

# Função de 3 imagens

```python
def plot_3img(img1,tit1,img2,tit2,img3,tit3,tam):
    plt.figure(figsize = tam)
    
    plt.subplot(1,3,1)
    plt.imshow(img1)
    plt.title(tit1)
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(1,3,2)
    plt.imshow(img2)
    plt.title(tit2)
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(1,3,3)
    plt.imshow(img3)
    plt.title(tit3)
    plt.xticks([]), plt.yticks([])
    
    plt.show()
```

# Funções para transformar BGE para RBG usando openCV

```python
def cv_mostrar_imagem_2(img,title,img2,title2,tam):
    img =  cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img2 =  cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
    
    plt.figure(figsize = tam)
    
    plt.subplot(1,2,1)
    plt.imshow(img)
    plt.title(title)
    plt.xticks([]), plt.yticks([])

    
    plt.subplot(1,2,2)
    plt.imshow(img2)
    plt.title(title2)
    plt.xticks([]), plt.yticks([])

    plt.show()
```

```python
def cv_mostrar_imagem_3(img,title,img2,title2,img3,title3,tam):
    img =  cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img2 =  cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
    img3 =  cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
    
    plt.figure(figsize = tam)
    
    plt.subplot(1,3,1)
    plt.imshow(img)
    plt.title(title)
    plt.xticks([]), plt.yticks([])

    
    plt.subplot(1,3,1)
    plt.imshow(img2)
    plt.title(title2)
    plt.xticks([]), plt.yticks([])

    plt.subplot(1,3,1)
    plt.imshow(img3)
    plt.title(title3)
    plt.xticks([]), plt.yticks([])
    
    plt.show()
```
