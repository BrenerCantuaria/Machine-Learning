```python
import numpy as np
import pandas as pd

from sklearn import tree
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
breast_cancer = load_breast_cancer(as_frame = True)
df_breast = breast_cancer.frame
```

```python
df_breast
```

```python
X_breast_cancer = df_breast.drop('target',axis = 1)
Y_breast_cancer = df_breast['target']
```

```python
X_breast_cancer.shape, Y_breast_cancer.shape
```

```python
X_train_breast_cancer,X_test_breast_cancer,Y_train_breast_cancer,Y_test_breast_cancer = train_test_split(X_breast_cancer,Y_breast_cancer,test_size = 0.3, random_state = 0)
```

```python
X_train_breast_cancer.shape, X_test_breast_cancer.shape, Y_train_breast_cancer.shape, Y_test_breast_cancer.shape
```

# Modelo Tree

```python
modelo_tree = DecisionTreeClassifier(random_state = 0)
```

```python
modelo_tree.fit(X_train_breast_cancer,Y_train_breast_cancer)
```

# Predição 

```python
y_preds_breast_cancer = modelo_tree.predict(X_test_breast_cancer)
```

```python
acuraria_tree = accuracy_score(Y_test_breast_cancer,y_preds_breast_cancer)
```

```python
print('Acurácia árvore de decisão: ', acuraria_tree)
```

# Relatário de classificação

```python
print(classification_report(Y_test_breast_cancer,y_preds_breast_cancer))
```

# Matriz de confusão: 

```python
print(confusion_matrix(Y_test_breast_cancer, y_preds_breast_cancer))
```

# Modelo RandomForest

```python
X_train_breast_cancer,X_test_breast_cancer,Y_train_breast_cancer,Y_test_breast_cancer = train_test_split(X_breast_cancer,Y_breast_cancer,test_size = 0.20, random_state = 0)
```

```python
X_train_breast_cancer.shape, X_test_breast_cancer.shape, Y_train_breast_cancer.shape, Y_test_breast_cancer.shape
```

```python
acuraria_rf = accuracy_score(Y_test_breast_cancer,y_preds_breast_cancer_rf)
```

# Acurácia RandomForest

```python
print('Acurácia Random Forest: ', acuraria_rf)
```
